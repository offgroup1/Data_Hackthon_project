```python
data['coupon_discount_amount'] = data['sale_price'] - data['transaction_amount']#coupon_discount결측치 제거
```

```python
data['sale_price'] = data['sale_price'].fillna(data['transaction_amount']) #sale_price 결측치 제거
```

```python
#유입, 유입 요청, 이탈를 카테고리 타이틀을 기준으로 한 함수입니다.
#inflow : 구매 완료자7


#inflow1: 패캠에서 구매 요청
#outflow: 이탈자.환불
#

def class_customer(v1):

    inflow = len(data[(data['type'] == 'PAYMENT') & (data['state'] == 'COMPLETED') & (data['category_title'] == v1)])
    inflow1 = len(data[(data['type'] == 'PAYMENT') & (data['state'] == 'REQUESTED') & (data['category_title'] == v1)])
    outflow = len(data[(data['type'] == 'REFUND') & (data['state'] == 'COMPLETED') & (data['category_title'] == v1)])
    return inflow , inflow1, outflow

```

```python
a = []
a = class_customer('프로그래밍')

print('(유입,유입 요청, 이탈)',a)
```

```python
a = class_customer('데이터사이언스')
print('(유입,유입 요청, 이탈)',a)
```

```python
a = class_customer('업무 생산성')

print('(유입,유입 요청, 이탈)',a)
```
